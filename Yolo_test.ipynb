```python
import cv2
import torch

# 1. Load a YOLO model (here using YOLOv5 from Ultralytics)
# Make sure you have torch and opencv-python installed (`pip install torch opencv-python`)
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# 2. Set your stream URL (adjust port/path as needed)
stream_url = 'http://192.168.0.148/'  # Replace <ESP32_IP> with your ESP32-CAM's IP

# 3. Open the stream
cap = cv2.VideoCapture(stream_url)

if not cap.isOpened():
    print("Error: Could not open video stream from ESP32-CAM.")
    exit()

while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab a frame")
        continue

    # 4. Run YOLO object detection on the frame
    results = model(frame)

    # 5. Render results on the frame (results.render() returns a list of annotated images)
    annotated_frame = results.render()[0]

    # 6. Display the annotated frame
    cv2.imshow('YOLO Object Detection', annotated_frame)
    
    # Exit if 'q' is pressed
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Cleanup
cap.release()
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np
import requests
import torch

# Load the YOLO model
model = torch.hub.load('ultralytics/yolov5', 'yolov5s', pretrained=True)

# Set your stream URL
stream_url = 'http://192.168.0.148'

# Start the stream using requests
stream = requests.get(stream_url, stream=True)
bytes_data = b''

for chunk in stream.iter_content(chunk_size=1024):
    bytes_data += chunk
    # Look for JPEG start and end markers
    start = bytes_data.find(b'\xff\xd8')
    end = bytes_data.find(b'\xff\xd9')
    if start != -1 and end != -1:
        jpg = bytes_data[start:end+2]
        bytes_data = bytes_data[end+2:]
        # Decode the image
        frame = cv2.imdecode(np.frombuffer(jpg, dtype=np.uint8), cv2.IMREAD_COLOR)
        if frame is None:
            continue

        # Run YOLO object detection
        results = model(frame)
        annotated_frame = results.render()[0]

        # Display the annotated frame
        cv2.imshow('YOLO Object Detection', annotated_frame)
        if cv2.waitKey(1) & 0xFF == ord('q'):
            break

cv2.destroyAllWindows()
```
